**Detalle de DataFrames:**

```python
# Verificar si los id_sucursal en df_vendedores existen en los id_sucursal de df_sucursales
ids_sucursal_vendedores = df_vendedores['id_sucursal'].unique()
ids_sucursal_sucursales = df_sucursales['id_sucursal'].unique()

# Encontrar id_sucursal en df_vendedores que no tienen una sucursal asociada en df_sucursales
ids_sucursal_sin_correspondencia = [id_sucursal for id_sucursal in ids_sucursal_vendedores if id_sucursal not in ids_sucursal_sucursales]

if ids_sucursal_sin_correspondencia:
    print(f"Los siguientes id_sucursal en df_vendedores no tienen una sucursal asociada en df_sucursales: {ids_sucursal_sin_correspondencia}")
else:
    print("Todos los id_sucursal en df_vendedores tienen una sucursal asociada en df_sucursales.")
```

```python
# Verificar si las ciudades en df_clientes existen en las ciudades de df_sucursales
ciudades_clientes = df_clientes['ciudad'].unique()
ciudades_sucursales = df_sucursales['ciudad'].unique()

# Encontrar ciudades en df_clientes que no tienen una sucursal asociada
ciudades_sin_sucursal = [ciudad for ciudad in ciudades_clientes if ciudad not in ciudades_sucursales]

if ciudades_sin_sucursal:
    print(f"Las siguientes ciudades en df_clientes no tienen una sucursal asociada en df_sucursales: {ciudades_sin_sucursal}")
else:
    print("Todas las ciudades en df_clientes tienen una sucursal asociada en df_sucursales.")
```

```python
# Diccionario para mapear nombres de productos a categorías y subcategorías corregidas
correcciones_productos = {
    'Coca Cola 1.5L': {'categoria': 'Bebidas', 'subcategoria': 'Gaseosas'},
    'Pepsi 1.5L': {'categoria': 'Bebidas', 'subcategoria': 'Gaseosas'},
    'Sprite 1.5L': {'categoria': 'Bebidas', 'subcategoria': 'Gaseosas'},
    'Fanta Naranja 1.5L': {'categoria': 'Bebidas', 'subcategoria': 'Gaseosas'},
    'Agua Mineral 500ml': {'categoria': 'Bebidas', 'subcategoria': 'Aguas'},
    'Jugo de Naranja 1L': {'categoria': 'Bebidas', 'subcategoria': 'Jugos'},
    'Jugo de Manzana 1L': {'categoria': 'Bebidas', 'subcategoria': 'Jugos'},
    'Energética Nitro 500ml': {'categoria': 'Bebidas', 'subcategoria': 'Energéticas'},
    'Yerba Mate Suave 1kg': {'categoria': 'Infusiones', 'subcategoria': 'Yerba Mate'},
    'Yerba Mate Intensa 1kg': {'categoria': 'Infusiones', 'subcategoria': 'Yerba Mate'},
    # Agregar aquí más correcciones según sea necesario
}

# Aplicar las correcciones al DataFrame df_productos
for nombre_producto, correccion in correcciones_productos.items():
    df_productos.loc[df_productos['nombre_producto'] == nombre_producto, 'categoria'] = correccion['categoria']
    df_productos.loc[df_productos['nombre_producto'] == nombre_producto, 'subcategoria'] = correccion['subcategoria']

# Verificar las correcciones mostrando los productos modificados
productos_modificados = df_productos[df_productos['nombre_producto'].isin(correcciones_productos.keys())]
display(productos_modificados)

# Mostrar los valores únicos de las columnas categoria y subcategoria después de las correcciones
print("\nValores únicos de la columna 'categoria' después de corregir:")
display(df_productos['categoria'].unique())

print("\nValores únicos de la columna 'subcategoria' después de corregir:")
display(df_productos['subcategoria'].unique())
```
